```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data.drop(['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
data
```

```python
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo= OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
geo_encoder.toarray()
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data=pd.concat([data.drop('Geography', axis=1),geo_encoded_df], axis=1)
data.head()
```

```python
##Save the encoders and scaler
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo, file)

```

```python
x= data.drop('Exited',axis=1)
y= data['Exited']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train =scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)

```

# ANN Implementation

ANN is called sequential model in Tensorflow

Keras-->Dense library or class (if we want to create Hidden neurons)

Key parameters used:
1. Initialise sequential network (while creating ANN)
2. Create a hidden neuron(nodes inside hidden layers) (not hidden layer)--->use Dense
3. Activation function(Sigmoid, tanh, ReLu, Leaky Relu). Generally advised to use ReLu in hidden layers & Sigmoid/Softmax in Output layer.
4. Optimizers----> USed in Back propagation forupdating the weights.
5. Loss functins--->
6. Metrics--> for Classification we use Accuracy & for Regression we use any of MSE, MAE etc
7. Training info --> store logs in a folder--> so that I can use Tensorboard library-->Its main aim is to display this logs that we can understand (Visualization)

Step wise process:
1. Created the model(Build ANN model)
2. Import required libraries
3. Setup optimizers and loss functions
4. Assign those optimizer and loss to the model along with metrics.
5. Setup the tensorboard, create required logs (required while training) &(used for visualisation) and create required call backs
6. Train the model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
#Build ANN
#1. Create Sequential model

model= Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),  #Hidden layer 1 connected with input layer(Dense for hiddenlayer & activation for activation function & input_shape for connecting input to the hidden layer(only for the first hidden layer, we need to write the input_shape))
    Dense(32,activation='relu'), ##HL2
    Dense(1,activation='sigmoid') ##Output layer
]
)
```

```python
model.summary()
```

```python
#Another way of applying optimizers
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
#Adam has a fixed learning rate. If you want to change you can do it.

#Can set loss also
loss= tensorflow.keras.losses.BinaryCrossentropy()
```

```python
#Compile the model to do Forward and backward propagation
#model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy']) 
# #If Multi classification,use 'sparse entropy'(can check in keras) in the loss above
# Can use optimizer='adam'(whatever optimizer or loss you need ) directly or can define an optimizer/loss and then give the details here like shown below:
model.compile(optimizer=opt,loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#Setup tensorboard 

from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

log_dir = 'logs/fit/' + datetime.datetime.now().strftime('%Y%MD-%H%M%S')

tensorflow_callback= TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#Setup Early stopping

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True) #val_loss=Validation loss, patience=no.of Epochs
```

```python
#Train the model

history= model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,   #y_test data will be used for comparision wrt accuracy
    callbacks=[tensorflow_callback,early_stopping_callback]
)

#After training, we will load the whole tensorboard
```

```python
#save this model file

model.save('model.h5') #h5 file will be compatible with keras
```
